**3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame**

**a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.**

**b) Top 10 ODI Batsmen along with the records of their team and rating.**

**c) Top 10 ODI bowlers along with the records of their team and rating.**

**Answer:-**

**Answer of a)**

```python
from bs4 import BeautifulSoup
import requests

url="https://www.icc-cricket.com/rankings/mens/team-rankings/odi"

code=requests.get(url)
if response.status_code==200:
    soup=BeautifulSoup(code.content)
    Teams=[]
    Match=[]
    Points=[]
    Ratings=[]
    
    for T in soup.find_all("span",class_="u-hide-phablet"):
        Teams.append(T.text)
        
     
    for M in soup.find_all("td",class_="table-body__cell u-center-text"):
        Match.append(M.text)
Matches=Match[::2]
Points=Match[1::2]
#print(Matches)
#print(Points)
ind=soup.find("td",class_="rankings-block__banner--matches")
Matches.insert(0,ind.text)
#print(Matches)

ind_match_total=soup.find("td",class_="rankings-block__banner--points")

Points.insert(0,ind_match_total.text)
#print(Points)


first_ele_of_Rating_tag=soup.find("td",class_="rankings-block__banner--rating u-text-right")
first_ele_text=first_ele_of_Rating_tag.text
import re
pattern="\d+"
Target=first_ele_text

only_rating=re.findall(pattern,Target)

for R in soup.find_all("td",class_="table-body__cell u-text-right rating"):
    Ratings.append(R.text)
Ratings.insert(0,only_rating[0])
#print(Ratings)


df=pd.DataFrame({"Teams":Teams[:10],"Matches":Matches[:10],"Points":Points[:10],"Ratings":Ratings[:10]})
df
```

```python
code=requests.get("https://www.cnbc.com/world/?region=world")

if code.status_code==200:
    soup=BeautifulSoup(code.content)
    Headlines=[]
    for lines in soup.find_all("div",class_="RiverHeadline-headline RiverHeadline-hasThumbnail"):
        Headlines.append(lines.text)
news=pd.DataFrame({"Headlines":Headlines})        
print(news)
```

**7) Write a python program to scrape mentioned details from dineout.co.in and make data frame.**

**i) Restaurant name**

**ii) Cuisine**

**iii) Location**

**iv) Ratings**

**v) Image URL**

**Answer of i) Restaurant name**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

code=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
if code.status_code==200:
    soup=BeautifulSoup(code.content)
    Restaurant=[]
    for i in soup.find_all("div",class_="restnt-info cursor"):
        Restaurant.append(i.text)
Hotel=pd.DataFrame({"Restaurant name":Restaurant})    
Hotel
        
```

**Answer of ii) Cuisine**

```python
code=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
if code.status_code==200:
    soup=BeautifulSoup(code.content)
    Cuisine=[]
    for i in soup.find_all("span",class_="double-line-ellipsis"):
        Cuisine.append(i.text)
Cuisine=pd.DataFrame({"Cuisine":Cuisine})    
Cuisine
```

**Answer of iii) Location**

```python
Location=[]
for i in soup.find_all("div",class_="restnt-loc ellipsis"):
    Location.append(i.text)
loc=pd.DataFrame({"Location":Location})    
loc
```

**Answer of iv) Ratings**

```python
Rating=[]

for i in soup.find_all("div",class_="restnt-rating rating-4"):
    Rating.append(i.text)
Rat=pd.DataFrame({"Rating":Rating})
Rat
```

**Answer of v) Image URL**

```python
image=[]

for i in soup.find_all("img",class_="no-img"):
    image.append(i["data-src"])
img=pd.DataFrame({"Image URL":image})
img
```

```python
df=pd.concat([Hotel,Cuisine,loc,Rat,img],axis=1,names=["Restaurant name","Cuisine","Location","Ratings","Image URL"])

df
```
