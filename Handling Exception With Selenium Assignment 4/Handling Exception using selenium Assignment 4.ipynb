**1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:** 

**A) Rank** 

**B) Name** 

**C) Artist** 

**D) Upload date** 

**E) Views**

**7. Scrape the details most watched tv series of all time from imdb.com.** 

**Url = https://www.imdb.com/list/ls095964455/ You have to find the following details:** 

**A) Name** 

**B) Year span** 

**C) Genre** 

**D) Run time**

**E) Ratings** 

**F) Votes**

**Answer:-**

```python
import re
```

```python
driver=webdriver.Chrome()
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
Name=[]
Year=[]
Genre=[]
RunTime=[]
Ratting=[]
Votes=[]
```

```python
try:
    name_tag=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]')
    for n in name_tag:
        n_text=n.text
        Name.append(n_text)
        
except NoSuchElementException:
    Name.append("--")
  
try:
    year_tag=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
    for y in year_tag:
        y_text=y.text
        Year.append(y_text)
        
except NoSuchElementException:
    Year.append("--")
    
    
try:
    genre_tag=driver.find_elements(By.XPATH,'//span[@class="genre"]')
    for g in genre_tag:
        g_text=g.text
        Genre.append(g_text)
        
except NoSuchElementException:
    Genre.append("--")
    
    
try:
    runtime_tag=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
    for r in runtime_tag:
        r_text=r.text
        RunTime.append(r_text)
        
except NoSuchElementException:
    RunTime.append("--")
    
    
try:
    ratting_tag=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-widget"]/div/span')
    for ra in ratting_tag:
        ra_text=ra.text
        Ratting.append(ra_text)
        
except NoSuchElementException:
    Ratting.append("--")
    
    
try:
    vote_tag=driver.find_elements(By.XPATH,'//span[@name="nv"]')
    for v in vote_tag:
        v_text=v.text
        Votes.append(v_text)
        
except NoSuchElementException:
    Votes.append("--")
```

```python
print(len(Name),len(Year),len(Genre),len(RunTime),len(Ratting),len(Votes))
```

```python
#Checking purpose
Name[:5]
```

```python
#Using regex to extract only the series name 
series_name=[]
for i in Name:
    pattern=r"\d+\.\s(.+?)\s*\("
    new_name=re.match(pattern,i)
    series_name.append(new_name.group(1))
    
```

```python
#Checking purpose
series_name[:5]
```

```python
print(len(series_name),len(Year),len(Genre),len(RunTime),len(Ratting),len(Votes))
```

```python
Populer_Series=pd.DataFrame({"Series Name":series_name,"Year Span":Year,"Genre":Genre,"Run Time":RunTime,"Rattings":Ratting[1::2],"Votes":Votes},index=(range(1,101)))
Populer_Series
```

**8. Details of Datasets from UCI machine learning repositories.**

**Url = https://archive.ics.uci.edu/** 

**You have to find the following details:** 

**A) Dataset name** 

**B) Data type** 

**C) Task** 

**D) Attribute type** 

**E) No of instances** 

**F) No of attribute** 

**G) Year** 
 
Note: - from the home page you have to go to the Show All Dataset page through code. 

**Answer:-**

```python
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
```

```python
#Clicking on View Dataset option
view_button=driver.find_element(By.XPATH,'//div[@class="flex flex-wrap justify-center gap-5"]/a[1]')
view_button.click()
```

```python
Dataset_name=[]
Task_Data_type_instances_attribute_no=[]
Attribute_type=[]
Year=[]
```

```python
#Scraping 11 pages only
start=0
end=11

for page in range(start,end):
    #Clicking on expand button and handling the ElementClickInterceptedException error
    try:
        expand=driver.find_element(By.XPATH,'//span[@class="swap-on text-primary-content"]')
        expand.click()
        
    except ElementClickInterceptedException:
        pass
    
    
    try:
        data_tag=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a')
        for n in data_tag:
            n_text=n.text
            Dataset_name.append(n_text)
    except NoSuchElementException:
        Dataset_name.append("--")
        
        
    try:
        task_d_type_instance_and_NoOfAttributes=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"]')
        for d in task_d_type_instance_and_NoOfAttributes:
            d_text=d.text
            Task_Data_type_instances_attribute_no.append(d_text)
    except NoSuchElementException:
        Task_Data_type_instances_attribute_no.append("--")  
        
        
    try:
        att_type=driver.find_elements(By.XPATH,'//tr/td[2]')
        for a in att_type:
            a_text=a.text
            Attribute_type.append(a_text)
    except NoSuchElementException:
        Attribute_type.append("--")
        
    try:
        year=driver.find_elements(By.XPATH,'//tr/td[3]')
        for y in year:
            y_text=y.text
            Year.append(y_text)
    except NoSuchElementException:
        Year.append("--")   
        
    #clicking on next button   
    next_button=driver.find_elements(By.XPATH,'//button[@class="btn-primary btn-sm btn"][2]')
    time.sleep(3)
        
```

```python
print(len(Dataset_name),len(Task_Data_type_instances_attribute_no),len(Attribute_type),len(Year))
```

```python
Dataset=pd.DataFrame({"Dataset name":Dataset_name,"Data Type":Task_Data_type_instances_attribute_no[1::4],"Task":Task_Data_type_instances_attribute_no[0::4],"Attribute Type":Attribute_type,"No Of Instances":Task_Data_type_instances_attribute_no[2::4],"No Of Attribute":Task_Data_type_instances_attribute_no[3::4],"Year":Year},index=range(1,111))
Dataset
```
