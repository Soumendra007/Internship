**5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.**

**Answer:-**

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.google.com/maps")
time.sleep(3)

location_name=input("Enter the location name")
search_bar=driver.find_element(By.XPATH,"/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input")
search_bar.clear()                             #clearing the searchbar
search_bar.send_keys(location_name)

search_button=driver.find_element(By.XPATH,"/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button")
search_button.click()

try:
    url=driver.current_url         #fetching current url open in Chrome 
    print("Your location url is :",url)
    latitude_longitude=re.findall(r"@(\d+\.\d+),(\d+\.\d+)",url)


    if latitude_longitude:
        for latitude,longitude in latitude_longitude:
            print("latitude = {} and longitude = {}".format(latitude,longitude))
    
    else:
        print("Unable to find Latitude and Longitude")
        
except NoSuchElementException:
    print("Error")
    
driver.close()    
```

**7. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”,“Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.**

**Answer:-**

```python
driver=webdriver.Chrome()
driver.get("https://www.forbes.com/billionaires/")
time.sleep(5)

Rank=[]
Name=[]
Net_worth=[]
Age=[]
Citizenship=[]
Source=[]
Industry=[]

try:
    rank_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][1]')
    for i in rank_tag:
        rank_text=i.text
        Rank.append(rank_text)
    
    Name_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][2]')
    for p in Name_tag:
        name_text=p.text
        Name.append(name_text)
        
    Net_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][3]')
    for n in Net_tag:
        net_text=n.text
        Net_worth.append(net_text)
        
    Age_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][4]')
    for a in Age_tag:
        age_text=a.text
        Age.append(age_text)
     
    Citizenship_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][5]')
    for c in Citizenship_tag:
        city_text=c.text
        Citizenship.append(city_text)
    
    Source_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][6]')
    for s in Source_tag:
        Source_text=s.text
        Source.append(Source_text)
        
    Industry_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][7]')
    for I in Industry_tag:
        Industry_text=I.text
        Industry.append(Industry_text)
    
except NoSuchElementException:
    Rank.append("--")
    Name.append("--")
    Net_worth.append("--")
    Age.append("--")
    Citizenship.append("--")
    Source.append("--")
    Industry.append("--")
    

billionaires=pd.DataFrame({"Rank":Rank,"Name":Name,"Net_worth":Net_worth,"Age":Age,"Citizenship":Citizenship,"Source":Source,"Industry":Industry})   
billionaires    
    
driver.close()    
    
    
    
    
```
