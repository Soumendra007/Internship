**Q1: Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.**

This task will be done in following steps:

        1. First get the webpage https://www.shine.com/

        2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.

        3. Then click the searchbutton.

        4. Then scrape the data for the first 10 jobs results you get.

        5. Finally create a dataframe of the scraped data.


**Answer:-**

```python
#importing required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
driver=webdriver.Chrome()
```

```python
#get the webpage
driver.get("https://www.shine.com/")
```

```python
#Entering ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field
job_title=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input")
job_title.send_keys("Data Analyst")
```

```python
#Entering ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.
#I could use here By.CLASS_NAME but rather than I chose absolute XPATH both can be usable because I just need to extract single element
loc=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
loc.send_keys("Bangalore")
```

```python
#Click on the search button
# using Absolute XPATH because have to extract one element
button=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
button.click()
```

```python
Job_title=[] 
job_location=[] 
company_name=[]
experience_required=[]
```

```python
#Scraping the job_titles
job=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')
for i in job:
    title=i.text
    Job_title.append(title)
    
#Scraping the job_location  
loca=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for l in loca:
    jobl=l.text
    job_location.append(jobl)  
    
#Scraping the company_name   
com=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for c in com:
    com_tag=c.text
    company_name.append(com_tag)
    
#Scraping the experience_required
exp=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for e in exp:
    exp_tag=e.text
    experience_required.append(exp_tag)
    
    
```

```python
#Checking every list lenth
print(len(Job_title),len(job_location),len(company_name),len(experience_required))
```

```python
Data_Analyst=pd.DataFrame({"Job Title":Job_title[:10],"Job Location":job_location[:10],"Company Name":company_name[:10],"Experience Required":experience_required[:10]},index=range(1,11))
Data_Analyst
```

```python
driver.close()
```

**Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:**

**6. Brand**

1. Go to Flipkart webpage by url :https://www.flipkart.com/
2. Enter ‚Äúsunglasses‚Äù in the search fieldwhere ‚Äúsearch for products, brands and more‚Äù is written and
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the
required data as usual.
4. After scraping data from the first page, go to the ‚ÄúNext‚Äù Button at the bottom other page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100sunglasses

**Answer:-**

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/")
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/div/input')
search.send_keys("sunglasses")
```

```python
#Clicking on search button
button=driver.find_element(By.XPATH,'//button[@class="_2iLD__"]')
button.click()
```

```python
brand=[]
product_description=[]
price=[]
offer=[]
```

```python
start=0
end=3

for page in range(start,end):
    brand_tag=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    product_description_tag_1=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    product_description_tag_2=driver.find_elements(By.XPATH,'//a[@class="IRpwTa _2-ICcC"]')
    price_tag=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    offer_tag=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]/span')
    for b in brand_tag:
        b_tag=b.text
        brand.append(b_tag)
        if len(brand)==100:
            break
    for p in product_description_tag_1:
        p_tag=p.text
        product_description.append(p_tag)
        
        if len(product_description)==100:
            break
    
    for pri in price_tag:
        pri_tag=pri.text
        price.append(pri_tag)
        if len(price)==100:
            break
    for o in offer_tag:
        o_tag=o.text
        offer.append(o_tag)
        if len(offer)==100:
            break
    for pt in product_description_tag_2:
        
        ps=pt.text
        product_description.append(ps)
        if len(product_description)==100:
            break
    
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]/span')
    next_button.click()
    time.sleep(10)
        
    
```

```python
product=pd.DataFrame({"Brand Name":brand,"Product Details":product_description[:100],"Price":price,"Offer":offer},index=range(1,101))
product
```

```python
driver.quit()
```

```python
Quote=[]
Author=[]
Type_Of_Quotes=[]
```

```python
start=0
end=10

for page in range(start,end):
    quote_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
    author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    type_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    
    
    for q in quote_tag:
        q_tag=q.text
        Quote.append(q_tag)
        
    for a in author_tag:
        a_tag=a.text
        Author.append(a_tag)
       
    
    for t in type_tag:
        t_tag=t.text
        Type_Of_Quotes.append(t_tag)
        
   
    
        
    
    next_button=driver.find_element(By.XPATH,'//li[@class="next"]/a[1]')
    next_button.click()
    time.sleep(3)
    
```
